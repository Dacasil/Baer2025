```python
import base64
import json
import os

import requests
from dotenv import load_dotenv

# Lade Umgebungsvariablen aus der .env-Datei
load_dotenv()

# URL der API-Endpunkte
URL = "https://hackathon-api.mlo.sehlat.io"
start_game_url = f"{URL}/game/start"
decision_url = f"{URL}/game/decision"

# Spielername
player_name = os.getenv("PLAYER_NAME")
api_key = os.getenv("API_KEY")
```

```python
headers = {
    "x-api-key": api_key,
    "Content-Type": "application/json"
}

start_response = requests.post(start_game_url, headers=headers, json={
    "player_name": player_name
})

if start_response.status_code == 200:
    start_data = start_response.json()
    session_id = start_data["session_id"]
    client_id = start_data["client_id"]
    print(f"Spiel erfolgreich gestartet! Session ID: {session_id}")

    # Speichere die Start-Response als .json
    with open("../data/_start_response.json", "w") as f:
        json.dump(start_data, f)

else:
    print(f"Fehler beim Starten des Spiels: {start_response.status_code}")
    print(f"Response Text: {start_response.text}")
```

```python
from pprint import pprint
client_data = start_data["client_data"]
pprint(client_data)
```

```python
# Define file paths for saving the decoded data
output_files = {
    "account": "../data/account.pdf",
    "description": "../data/description.txt",
    "passport": "../data/passport.png",
    "profile": "../data/profile.docx"
}

# Decode and save each file
for key, file_path in output_files.items():
    encoded_data = client_data.get(key)
    if encoded_data:
        with open(file_path, "wb") as file:
            file.write(base64.b64decode(encoded_data))
        print(f"{key.capitalize()} saved to {file_path}")
    else:
        print(f"No data found for {key}")
```
